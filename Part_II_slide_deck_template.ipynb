## Investigation Overview


> In this investigation, I wanted to look at the dependence of distance travelled and time on bike sharing, type of user, gender, parts of the day, days of the week and generations. 


## Dataset Overview

> The data contained information regarding individual rides made in a bike-sharing system covering the greater San Francisco Bay area. It contained information including duration of ride, start and end time of ride, where the ride started and ended, type of user, year of birth, gender, and whether users share bike for trips.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
bikedata2 = pd.read_csv('bikedata2.csv')
```

```python
#group dataset by generations
group_gen = bikedata2.groupby('Gen')
#get generations by groups
baby_boomers = group_gen.get_group('Baby Boomers')
gen_x = group_gen.get_group('Gen X')
gen_z = group_gen.get_group('Gen Z')
greatest_gen = group_gen.get_group('Greatest Gen')
lost_gen = group_gen.get_group('Lost Gen')
millen = group_gen.get_group('Millenials')
silent_gen = group_gen.get_group('Silent Gen')
```

## Average Distance and Time for Bike Sharing

> Regarding the possibility of average distance of travel and time spent on trips being dependent on whether a user shares their bike or not, those who do not share their bikes spent more time and travelled more distance. 

```python
ax = bikedata2.groupby('bike_share_for_all_trip').mean().round(2)[['time_hour', 'distance_km']
                                          ].plot(kind='bar', stacked=True, 
                                                 title='Average Time and Distance Based on Bike Sharing', fontsize=9);
ax.set_xlabel("Bike Sharing", fontsize=12)
ax.set_ylabel("Average Value", fontsize=12)
plt.legend(["Time(Hour)", "Distance(Km)"], loc ="upper right");
```

## Average Distance and Time for User Types

> When it comes to the possibility of average distance of travel and time spent on trips being dependent on whether a user is a customer or subscriber, customers spent more time and travelled more distance. 

```python
ax = bikedata2.groupby('user_type').mean().round(2)[['time_hour', 'distance_km']].plot(
    kind='bar', title='Time and Distance Based on User Type', stacked=True);
ax.set_xlabel("User Type", fontsize=12)
ax.set_ylabel("Average Value", fontsize=12)
plt.legend(["Time(Hour)", "Distance(Km)"], loc ="upper right");
```

## Average Distance and Time for Genders

> Regarding the possibility of average distance of travel and time spent on trips being dependent on the gender of a user, users who did not specify their genders spent more time on the average, while users who are neither male nor female travelled the longest distance. Females spent more time and travelled longer distances than males.

```python
ax = bikedata2.groupby('member_gender').mean().round(2)[['time_hour', 'distance_km']
                                         ].plot(kind='bar', stacked=True, 
                                                title='Average Time and Distance for Genders', 
                                                figsize= (6,4));
ax.set_xlabel("Gender", fontsize=12)
ax.set_ylabel("Average Value", fontsize=12)
plt.legend(["Time(Hour)", "Distance(Km)"], bbox_to_anchor=(1.05, 1.0), loc='upper left');
```

## Average Distance and Time for Generations

> In terms of average distance of travel and time spent on trips being dependent on the generation a user belongs to, baby boomers spent more time on trips than other generations while lost generation travelled longest distance. 

```python
ax = bikedata2.groupby('Gen').mean().round(2)[['time_hour', 'distance_km']
                                         ].plot(kind='bar', stacked=True, 
                                                title='Average Time and Distance for Genders', 
                                                figsize= (6,4));
ax.set_xlabel("Generation", fontsize=12)
ax.set_ylabel("Average Value", fontsize=12)
plt.legend(["Time(Hour)", "Distance(Km)"], bbox_to_anchor=(1.05, 1.0), loc='upper left');
```

## Average Distance and Time for Days of the Week

> Users who started their trip on Wednesday and ended it on Thursday travelled longest distance and also spent more time. 

```python
ax = bikedata2.groupby('start_end_days').mean().round(2)[['time_hour', 'distance_km']
                                          ].plot(kind='bar', stacked=True, figsize= (12,5), title='Time and Distance for Days');
ax.set_xlabel("Start and End Days", fontsize=12)
ax.set_ylabel("Average Value", fontsize=12)
plt.legend(["Time(Hour)", "Distance(Km)"], bbox_to_anchor=(1.0, 1.0), loc='upper left');
```

## Average Distance and Time for Parts of the Day

> Users who started their trip in the evening and ended it in the afternoon travelled the longest distance while those who started their trip in the morning and ended it in early morning spent more time. 

```python
ax = bikedata2.groupby('part_days').mean().round(2)[['time_hour', 'distance_km']].plot(kind='bar', stacked=True,
                                                                         title='Time and Distance for Parts of the Day', figsize=(12, 7));
ax.set_xlabel("Parts of the Day", fontsize=12)
ax.set_ylabel("Average Value", fontsize=12)
plt.legend(["Time(Hour)", "Distance(Km)"], bbox_to_anchor=(1.0, 1.0), loc='upper left');
```
